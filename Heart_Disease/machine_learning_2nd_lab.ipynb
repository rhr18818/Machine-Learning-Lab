```python
#scaling
from sklearn.preprocessing import StandardScaler

#Identify numeric column excluding the target
numeric_cols = df.select_dtypes(include=['number']).columns.tolist()
if 'targer' in numeric_cols:
    numeric_cols.remove('target')

#Apply StandardSclaer
scaler = StandardScaler()
df[numeric_cols] = scaler.fit_transform(df[numeric_cols])

#Save the processed DataFrame to a CSV file
df.to_csv('/content/drive/MyDrive/NWU/Machine-Learning-Lab/heart_disease_uci_processed.csv', index=False)
df.to_csv('heart_disease_uci_processed.csv', index=False)

#Show the first few Rows of the final DataFrame
df.head()
```
